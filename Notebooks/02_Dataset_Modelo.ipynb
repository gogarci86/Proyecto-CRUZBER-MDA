## Estandarizamos los nombres de las provincias para cruzar los dataframes


```python
set_agrupado = set(df_agrupado['Provincia'].unique())
set_ciclismo = set(df_ciclismo['Provincia'].unique())

# Provincias que están en ciclismo pero faltan en tu tabla agrupada
print("Faltan en df_agrupado:", set_ciclismo - set_agrupado)

# Provincias que sobran en agrupado o están escritas distinto
print("Faltan en df_ciclismo:", set_agrupado - set_ciclismo)
```

```python
import unicodedata

def limpiar_texto(texto):
    if pd.isna(texto): return texto
    # Normalizar para separar caracteres de tildes
    texto = unicodedata.normalize('NFD', str(texto))
    # Filtrar solo caracteres que no sean tildes, pasar a mayúsculas y quitar espacios
    return "".join(c for c in texto if unicodedata.category(c) != 'Mn').upper().strip()

# Aplicar limpieza a la columna Provincia en ambos DataFrames
df_agrupado['Provincia'] = df_agrupado['Provincia'].apply(limpiar_texto)
df_ciclismo['Provincia'] = df_ciclismo['Provincia'].apply(limpiar_texto)
```

```python
# Buscar filas en ciclismo cuya provincia no exista en el agrupado
invalidas = df_ciclismo[~df_ciclismo['Provincia'].isin(df_agrupado['Provincia'])]

if invalidas.empty:
    print("✅ ¡Perfecto! Todas las provincias de df_ciclismo existen en df_agrupado.")
else:
    print(f"⚠️ Atención: Hay {invalidas['Provincia'].nunique()} nombres que aún no coinciden.")
    print(invalidas['Provincia'].unique())
```

```python
# Definimos la traducción: 'Valor_en_ciclismo': 'Valor_en_agrupado'
mapeo_provincias = {
    'BALEARES': 'ILLES BALEARS',
    'S.C TENERIFE': 'STA CRUZ DE TENERIFE',
    'BIZKAIA': 'VIZCAYA'
}

# Aplicamos el cambio solo en df_ciclismo
df_ciclismo['Provincia'] = df_ciclismo['Provincia'].replace(mapeo_provincias)

# Volvemos a verificar
invalidas = df_ciclismo[~df_ciclismo['Provincia'].isin(df_agrupado['Provincia'])]

if invalidas.empty:
    print("✅ ¡Conseguido! Ahora todos los nombres coinciden.")
else:
    print(f"Aún quedan: {invalidas['Provincia'].unique()}")
```

## Leemos datos de Clima

```python
# Leer el dataset de clima
df_clima = pd.read_csv('../Datasets/clima_semanal_openmeteo.csv')

# 1. Filtrar el rango específico: 2022, 2023 y 2024
df_clima_filtrado = df_clima[df_clima['year'].isin([2022, 2023, 2024])].copy()

# 2. Estandarizar provincias y crear la llave de unión
# (Asegúrate de haber ejecutado la función limpiar_texto previamente)
df_clima_filtrado['provincia'] = df_clima_filtrado['provincia'].apply(limpiar_texto)
df_clima_filtrado['anio_semana'] = df_clima_filtrado['year'].astype(str) + '-' + df_clima_filtrado['semana'].astype(str)

# 3. Agrupar por semana y provincia con las métricas solicitadas
df_clima_agrupado = df_clima_filtrado.groupby(['anio_semana', 'provincia']).agg({
    'temp_media': 'mean',
    'precip_mm': 'mean',
    'viento_max': 'mean'
}).reset_index()

# 4. Renombrar para que el Merge sea limpio
df_clima_agrupado.columns = ['anio_semana', 'Provincia', 'temp_media_sem', 'precip_media_sem', 'viento_max_medio_sem']

# Verificación de los años resultantes
print(f"Años en el dataset de clima: {sorted(df_clima_filtrado['year'].unique())}")
display(df_clima_agrupado.head())
```

```python
# 1. Aseguramos que los nombres de las columnas en el filtrado sean los correctos
# Creamos 'Provincia' (mayúscula) a partir de 'provincia' (minúscula)
df_clima_filtrado['Provincia'] = df_clima_filtrado['provincia'].apply(limpiar_texto)

# 2. Ahora sí, agrupamos usando 'Provincia' con P mayúscula
df_clima_agrupado = df_clima_filtrado.groupby(['anio_semana', 'Provincia']).agg({
    'temp_media': 'mean',
    'precip_mm': 'mean',
    'viento_max': 'mean'
}).reset_index()

# 3. Renombramos las columnas finales
df_clima_agrupado.columns = [
    'anio_semana', 
    'Provincia', 
    'clima_temp_media', 
    'clima_precip_media', 
    'clima_viento_max_medio'
]

display(df_clima_agrupado.head())
```

```python
# 1. Diccionario para unificar nombres entre Clima y Ventas
mapeo_clima_nombres = {
    'GUIPUZCOA': 'GIPUZKOA',
    'SANTA CRUZ DE TENERIFE': 'STA CRUZ DE TENERIFE',
    'BIZKAIA':'VIZCAYA'
}

# 2. Limpieza profunda del Clima antes de agrupar
df_clima['Provincia'] = df_clima['provincia'].apply(limpiar_texto)
df_clima['Provincia'] = df_clima['Provincia'].replace(mapeo_clima_nombres)
df_clima['anio_semana'] = df_clima['year'].astype(str) + '-' + df_clima['semana'].astype(str)

# 3. Agrupación (Asegurando que no hay espacios en las llaves)
df_clima_agrupado = df_clima[df_clima['year'].isin([2022, 2023, 2024])].groupby(['anio_semana', 'Provincia']).agg({
    'temp_media': 'mean',
    'precip_mm': 'mean',
    'viento_max': 'mean'
}).reset_index()

df_clima_agrupado.columns = ['anio_semana', 'Provincia', 'clima_temp', 'clima_precip', 'clima_viento']

# 4. LIMPIEZA CRÍTICA: Quitar espacios en blanco invisibles en AMBAS tablas
df_final['Provincia'] = df_final['Provincia'].astype(str).str.strip()
df_final['anio_semana'] = df_final['anio_semana'].astype(str).str.strip()
df_clima_agrupado['Provincia'] = df_clima_agrupado['Provincia'].astype(str).str.strip()
df_clima_agrupado['anio_semana'] = df_clima_agrupado['anio_semana'].astype(str).str.strip()

# 5. MERGE FINAL
df_maestro = pd.merge(df_final, df_clima_agrupado, on=['anio_semana', 'Provincia'], how='left')

# Verificación
print(f"¿Quedan nulos en A CORUNA?: {df_maestro[df_maestro['Provincia']=='A CORUNA']['clima_temp'].isna().sum()}")
display(df_maestro.head())
```

```python
#1. Aseguramos la función de limpieza (por si acaso)
def limpiar_texto(texto):
    if pd.isna(texto): return texto
    texto = unicodedata.normalize('NFD', str(texto))
    return "".join(c for c in texto if unicodedata.category(c) != 'Mn').upper().strip()

# 2. Diccionario de traducción para que el clima hable el mismo idioma que ventas
mapeo_clima = {
    'GUIPUZCOA': 'GIPUZKOA',
    'SANTA CRUZ DE TENERIFE': 'STA CRUZ DE TENERIFE'
}

# 3. Preparar df_clima_agrupado desde cero con correcciones
df_clima_filt = df_clima[df_clima['year'].isin([2022, 2023, 2024])].copy()
df_clima_filt['Provincia'] = df_clima_filt['provincia'].apply(limpiar_texto).replace(mapeo_clima)
df_clima_filt['anio_semana'] = df_clima_filt['year'].astype(str) + '-' + df_clima_filt['semana'].astype(str)

df_clima_agrupado = df_clima_filt.groupby(['anio_semana', 'Provincia']).agg({
    'temp_media': 'mean',
    'precip_mm': 'mean',
    'viento_max': 'mean'
}).reset_index()

df_clima_agrupado.columns = ['anio_semana', 'Provincia', 'clima_temp', 'clima_precip', 'clima_viento']

# 4. LIMPIEZA DE SEGURIDAD: Quitamos cualquier espacio invisible en ambos DataFrames
df_final['Provincia'] = df_final['Provincia'].astype(str).str.strip()
df_final['anio_semana'] = df_final['anio_semana'].astype(str).str.strip()

df_clima_agrupado['Provincia'] = df_clima_agrupado['Provincia'].astype(str).str.strip()
df_clima_agrupado['anio_semana'] = df_clima_agrupado['anio_semana'].astype(str).str.strip()

# 5. EL MERGE FINAL
df_maestro = pd.merge(df_final, df_clima_agrupado, on=['anio_semana', 'Provincia'], how='left')

# Verificación de resultados
print(f"Total filas: {len(df_maestro)}")
print(f"Nulos en A CORUNA: {df_maestro[df_maestro['Provincia'] == 'A CORUNA']['clima_temp'].isna().sum()}")
display(df_maestro.head())
```

```python
# 1. Creamos un diccionario de "vecinos"
vecinos = {
    'VIZCAYA': 'GIPUZKOA',
    'CEUTA': 'CADIZ',
    'MELILLA': 'MALAGA'
}

# 2. Iteramos para rellenar los huecos
for prov_nula, prov_vecina in vecinos.items():
    # Extraemos el clima de la provincia vecina
    clima_vecino = df_clima_agrupado[df_clima_agrupado['Provincia'] == prov_vecina].copy()
    clima_vecino['Provincia'] = prov_nula # Le cambiamos el nombre para que coincida
    
    # Actualizamos los nulos en df_maestro usando los datos del vecino
    df_maestro.update(
        df_maestro[df_maestro['Provincia'] == prov_nula].merge(
            clima_vecino, on=['anio_semana', 'Provincia'], suffixes=('_old', '')
        ).set_index(df_maestro[df_maestro['Provincia'] == prov_nula].index)
    )

print("✅ Clima de Vizcaya, Ceuta y Melilla completado usando provincias vecinas.")
```
